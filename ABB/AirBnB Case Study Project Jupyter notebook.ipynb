```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

air_bnb = pd.read_csv('AB_NYC_2019.csv')
# air_bnb
# air_bnb.set_option('display.max_rows',100)
```

```python
# How many neighborhood groups are available and which shows up the most?

print('There are 5 Neighborhood groups and Manhattan shows up the most.')
air_bnb.groupby(['neighbourhood_group']).sum().sort_values('calculated_host_listings_count',ascending=False)
```

```python
# Are private rooms the most popular in manhattan?
print('Not at all, Entire home/apt is the most popular in room type in Manhattan. ')
air_bnb.groupby(['room_type','neighbourhood_group']).count().sort_values('room_type', ascending=False)

```

```python
# Which hosts are the busiest and based on their reviews?
print('Michael is the busiest based on reviews.')
air_bnb.groupby(['host_name']).sum().sort_values('number_of_reviews', ascending=False)
```

```python
#Which neighorhood group has the highest average price?
print('Manhattan has the highest average price of $196.88')

air_bnb.groupby('neighbourhood_group',as_index=False).mean().round(decimals=2)[['neighbourhood_group','price']]
```

```python
# Which neighbor hood group has the highest total price?
print('Manhattan has the highest total price.')
air_bnb.groupby(['neighbourhood_group']).sum()
```

```python
#Which top 5 hosts have the highest total price?
top5=air_bnb.groupby(['host_name']).sum().sort_values('price', ascending=False)
top5.head(5)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
# print(air_bnb.columns)
print('Michael has 0 availability with 2336 reviews')
air_bnb.groupby(['host_name','availability_365']).sum().sort_values('number_of_reviews', ascending=False)
```

```python
# What host has the highest total of prices and where are they located?
print(air_bnb.columns)
air_bnb.groupby(['host_name','neighbourhood']).sum().sort_values('price', ascending=False)
```

```python
# When did Danielle from Queens last receive a review?
hosts=air_bnb.groupby(['host_name','neighbourhood','neighbourhood_group']).count().sort_values('host_name',ascending=True)
hosts.head(60)
```
