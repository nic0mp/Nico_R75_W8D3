# What is the average weight of all players who played during this time?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

allStar = pd.read_csv('NBA All Stars 2000-2016 - Sheet1.csv')

print(allStar.columns)
allStar
```

```python
allStar.mean()[['WT']]
```

# What team is represented the most during this time?

```python
allStar.groupby(['Team']).count().sort_values('Year', ascending=False)
```

# What draft class is represeted the most during this time?

```python
print('1996 Draft Class has the most players appear in the All-Star')
dclass=allStar.groupby(['NBA Draft Status']).count().sort_values('Player', ascending=False)
dclass.head(30)
```

# Are foriegn players more prevalent in the All-Star Game during this time?

```python
pnationality=allStar.groupby(['Player','Nationality']).count().sort_values('Nationality', ascending=True)
pnationality.head(30)
```

# How often are the Western All-Stars voted in by fans?

```python
print('Fans voted for Western All-Stars 119 times')
confStars=allStar.groupby(['Selection Type']).count().sort_values('Selection Type', ascending=False)
confStars.head(50)
```

# How many times does Steph Curry make the All-Star game during this time period?

```python
print('Steph made the All-Star Team 4 times from 2009-2016')
steph=allStar.groupby(['Player']).count().sort_values('Selection Type', ascending=False)
steph.head(50)
```

# How many Shooting Gaurds have made the All-Star game during this time?

```python
print('57 SHooting Guards made the All-Star during this time')
sg=allStar.groupby(['Pos']).count().sort_values('Player', ascending=False)
sg.head(50)
```

# How many Chicago Bulls players have made the All-Star Game during this time period?

```python
print('Chicago Bulls players made the All-Star 12 times during this time')
bulls=allStar.groupby(['Team']).count().sort_values('Player', ascending=False)
bulls.head(33)
```
